绘制单调递增函数 $y=x$ 的图像

```python
import matplotlib.pyplot as plt

# Define the function
def f(x):
  return x

# Define the range of x-values
x = range(-10, 11)  # Adjust the range as needed

# Calculate the corresponding y-values
y = [f(i) for i in x]

# Create the plot
plt.plot(x, y, label="y = x")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = x")

# Add grid lines for better readability
plt.grid(True)

# Show the plot
plt.show()
```

绘制图像 单调递减函数 $y=-x$

```python
import matplotlib.pyplot as plt

# Define the function
def f(x):
  return -x

# Define the range of x-values
x = range(-10, 11)  # Adjust the range as needed

# Calculate the corresponding y-values
y = [f(i) for i in x]

# Create the plot
plt.plot(x, y, label="y = -x")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = -x")

# Add grid lines for better readability
plt.grid(True)

# Show the plot
plt.show()

print("The function y = -x is a monotonically decreasing function.")
```
