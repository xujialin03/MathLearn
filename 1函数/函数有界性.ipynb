绘制 $\frac{1}{x}$ 的图像，并画出在区间 (1,2)中的上下界

```python

import numpy as np
import matplotlib.pyplot as plt

# 定义x的范围
x = np.linspace(-10, 10, 400)
# 避免x为0导致的无穷大错误
x = x[x != 0]

# 计算1/x
y = 1 / x

# 创建图形
plt.figure(figsize=(8, 6))

# 绘制1/x的图像
plt.plot(x, y, label='y=1/x')

# 绘制y=1的直线
plt.axhline(y=1, color='r', linestyle='--', label='y=1')

# 绘制y=-1的直线
plt.axhline(y=-1, color='b', linestyle='--', label='y=-1')

# 添加标题和标签
plt.title('Graph of 1/x with y=1 and y=-1')
plt.xlabel('x')
plt.ylabel('y')

# 添加图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图像
plt.show()
```
