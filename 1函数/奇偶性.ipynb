绘制奇函数 $\ln({-x+\sqrt{1+x^2}})$ 的图像

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def f(x):
    return np.log(-x + np.sqrt(1 + x**2))

# 创建x值的数组
x = np.linspace(-100, 100, 400)

# 计算对应的y值
y = f(x)

# 创建图形
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$y = \ln(-x + \sqrt{1+x^2})$')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Plot of $y = \ln(-x + \sqrt{1+x^2})$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

奇函数 $f(x)=sinx$ 的图像

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the function
def f(x):
  return np.sin(x)

# Define the range of x-values
x = np.linspace(-math.pi, math.pi, 400)  # Adjust the range as needed

# Calculate the corresponding y-values
y = f(x)

# Create the plot
plt.plot(x, y, label="y = sin(x)")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = sin(x)")

# Add grid lines for better readability
plt.grid(True)

# Show the plot
plt.show()
```

可以看出奇函数关于 原点对称

绘制偶函数 $y=x^2$ 的图像 可以看出 偶函数关于 Y 轴对称

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义函数 y = x^2
def f(x):
    return x**2

# 生成 x 的值
x = np.linspace(-10, 10, 400)
y = f(x)

# 绘制函数图像
plt.plot(x, y, label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^2')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 判断是不是偶函数
def is_even_function(f, x_values):
    for x in x_values:
        if f(x) != f(-x):
            return False
    return True

# 生成测试的 x 值
test_x = np.linspace(-100, 100, 1000)

# 判断并打印结果
if is_even_function(f, test_x):
    print("函数 y = x^2 是偶函数")
else:
    print("函数 y = x^2 不是偶函数")
```

函数 $y=sinx+cosx $ 是非奇非偶函数

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the function
def f(x):
  return np.sin(x)+np.cos(x)

# Define the range of x-values
x = np.linspace(5*-math.pi, 5*math.pi, 400)  # Adjust the range as needed

# Calculate the corresponding y-values
y = f(x)

# Create the plot
plt.plot(x, y, label="y = sin(x)+cos(x)")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = sin(x)+cos(x)")

# Add grid lines for better readability
plt.grid(True)

# Show the plot
plt.show()
```
