周期性 有正数 $l$ 使得 $f(x+l)=f(x)$ 那么$f(x)$是周期函数，$l$称为 $f(x)$的周期 例如 $sin(x),cos(x)$ 都是以 $2\pi$ 为周期

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the function
def f(x):
  return np.sin(x)

# Define the range of x-values
x = np.linspace(2*-math.pi, 2*math.pi, 400)  # Adjust the range as needed

# Calculate the corresponding y-values
y = f(x)

# Create the plot
plt.plot(x, y, label="y = sin(x)")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = sin(x)")

# Add grid lines for better readability
plt.grid(True)

# Show the plot
plt.show()
```

函数 $y=tanx$ 是以 $\pi$ 为一个周期

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the function
def f(x):
  return np.tan(x)

# Define the range of x-values
x = np.linspace(-math.pi, math.pi, 400)  # Adjust the range as needed

# Calculate the corresponding y-values
y = f(x)

# Create the plot
plt.plot(x, y, label="y = tan(x)")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of y = tan(x)")

# Add grid lines for better readability
plt.grid(True)

# Show the plot
plt.show()
```
