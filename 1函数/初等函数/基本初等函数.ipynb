常数函数 $y=C$

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义 x 的范围
x = np.linspace(-10, 10, 400)

# 定义 y 的值
y = np.ones_like(x)

# 绘制图形
plt.plot(x, y, label='y = 1')

# 强制定义y的坐标最大值
plt.ylim(0,3)

# 添加标题和标签
plt.title('Plot of the Constant Function y = 1')
plt.xlabel('x')
plt.ylabel('y')

# 添加图例
plt.legend()

# 显示图形
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

幂函数 $y=x^u$

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function and the range of x
n=3
x = np.linspace(-10, 10, 400)
y = x**(n)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$y=x^{n}$', color='b')
plt.title(f'Plot of $y=x^{n}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

指数函数 $y=a^x (a>0且 a \neq 1)$
常用指数公式:  
$a^x*a^y=a^{xy}$  
$(a^x)^y=a^{xy}$  
$(ab)^x=a^{x}b^{x}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 2
x = np.linspace(-2, 2, 400)
y = a**x

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$y = {a}^x$')
plt.title(f'Exponential Function $y = {a}^x$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

对数函数 $y=log_a{x}$ 定义域为$(0，+\infty)$，值域也为$(-\infty，+\infty)$ 过定点(0,1)  
当 0<a<1 时,对数函数单调递减
当 a>1 时， 单调递增  

当 a 相同时 对数函数和指数函数互为么函数 $y=a^{x}$ -> $x=log_a{y}$  
以 e 为底 $y=lnx$ 以10为底$y=lgx$  
常用公式:  
$log_a{MN}=log_a{M}+log_a{N}$  
$log_a{\frac{M}{N}}=log_a{M}-log_a{N}$   
$log_a{M^N}=Nlog_a{M}$  
$log_aM=\frac{log_b{M}}{log_b{a}} (b>0 且 b\neq 1)$(换底公式) 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 定义 x 的范围，从 0.1 到 10，避免 x 为 0
x = np.linspace(0.1, 10, 400)
# 计算 y = ln(x)
y = np.log(x)

# 设置中文字体
# 指定字体路径，这里以SimHei为例
font_prop = fm.FontProperties(fname=font_path)

# 创建一个图形
plt.figure()

# 绘制 y = ln(x)
plt.plot(x, y, label='y = ln(x)')

# 添加标题和标签
plt.title('对数函数 y = ln(x)')
plt.xlabel('x')
plt.ylabel('y')

# 添加网格
plt.grid(True)

# 显示图例
plt.legend()

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置底数
a = 3

# 生成 x 值（从0.1到10，以避免对数函数在x=0处不定义）
x = np.linspace(0.1, 10, 400)
# 计算对应的 y 值
y = np.log(x) / np.log(a)

# 创建图像
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = \log_3(x)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Graph of $y = \log_3(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.ylim(-3, 3)  # 设定 y 轴范围
plt.show()
```
