```python
# 指定字体路径，这里以SimHei为例
font_path = 'C:/Windows/Fonts/simhei.ttf'  # 如果使用Windows系统可以改为对应的字体路径
```

双曲正弦： $sh x=\frac{e^x-e^{-x}}{2}$   定义域为$(-\infty，+\infty)$，值域也为$(-\infty，+\infty)$ 单调增加 奇函数

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 设置中文字体
# 指定字体路径，这里以SimHei为例
font_prop = fm.FontProperties(fname=font_path)

# 定义x的范围
x = np.linspace(-2, 2, 400)

# 定义双曲正弦函数
sinh = np.sinh(x)

# 定义渐近线 y = \frac{1}{2} e^x 和 y = -\frac{1}{2} e^{-x}
asymptote1 = 0.5 * np.exp(x)
asymptote2 = -0.5 * np.exp(-x)

# 创建图像
plt.figure(figsize=(10, 6))

# 绘制双曲正弦函数
plt.plot(x, sinh, label=r'$\sinh(x)$', color='blue')

# 绘制渐近线
plt.plot(x, asymptote1, label=r'$y = \frac{1}{2} e^x$', linestyle='--', color='red')
plt.plot(x, asymptote2, label=r'$y = -\frac{1}{2} e^{-x}$', linestyle='--', color='green')

# 添加标题和标签
plt.title('双曲正弦函数及其渐近线', fontproperties=font_prop)
plt.xlabel('x', fontproperties=font_prop)
plt.ylabel('y', fontproperties=font_prop)

# 添加图例
plt.legend(prop=font_prop)

# 显示图像
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

双曲余弦:    $ch x=\frac{e^x-e^{-x}}{2}$   定义域为$(-\infty，+\infty)$，值域也为$(1，+\infty)$ 单调增加 偶函数

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 设置中文字体
# 指定字体路径，这里以SimHei为例
font_prop = fm.FontProperties(fname=font_path)

# 定义x的范围
x = np.linspace(-2, 2, 400)

# 定义双曲余弦函数
cosh = np.cosh(x)

# 定义渐近线 y = 1/2 * e^x 和 y = 1/2 * e^-x
asymptote1 = 0.5 * np.exp(x)
asymptote2 = 0.5 * np.exp(-x)

# 创建图像
plt.figure(figsize=(10, 6))

# 绘制双曲余弦函数
plt.plot(x, cosh, label=r'$\cosh(x)$', color='blue')

# 绘制渐近线
plt.plot(x, asymptote1, label=r'$y = \frac{1}{2} e^x$', linestyle='--', color='red')
plt.plot(x, asymptote2, label=r'$y = \frac{1}{2} e^{-x}$', linestyle='--', color='green')

# 标记点 (0, 1)
plt.scatter(0, 1, color='black')  # 用黑色标记点
plt.annotate('(0, 1)', xy=(0, 1), xytext=(10, 10),
             textcoords='offset points', fontsize=12, color='black', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='black'))

plt.scatter(0, 0.5, color='red')  # 用红色记点
plt.annotate(r'(0, $\frac{1}{2}$)', xy=(0, 0.5), xytext=(10, 10),
             textcoords='offset points', fontsize=12, color='red', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='red'))

# 添加标题和标签
plt.title('双曲余弦函数及其渐近线', fontproperties=font_prop)
plt.xlabel('x', fontproperties=font_prop)
plt.ylabel('y', fontproperties=font_prop)

# 添加图例
plt.legend(prop=font_prop)

# 显示图像
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

```

双曲正切 $th x=\frac{sh x}{ch x}=\frac{e^{x}-e^{-x}}{e^x+e^{-x}}$  定义域为$(-\infty，+\infty)$，值域也为$(-1，1)$ 单调增加 奇函数

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义x的范围
x = np.linspace(-2, 2, 400)

# 计算tanh函数
y = np.tanh(x)

# 创建一个新的图形
plt.figure()

# 绘制tanh函数
plt.plot(x, y, label='tanh(x)')

# 绘制y=1的直线
plt.axhline(y=1, color='r', linestyle='--', label='y=1')

# 绘制y=-1的直线
plt.axhline(y=-1, color='b', linestyle='--', label='y=-1')

# 添加网格
plt.grid(True)

# 强调坐标线
plt.axhline(y=0, color='black', linewidth=1.5)
plt.axvline(x=0, color='black', linewidth=1.5)

# 添加标题和标签
plt.title('双曲正切')
plt.xlabel('x')
plt.ylabel('tanh(x)')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
