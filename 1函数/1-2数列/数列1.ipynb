绘制图像$\frac{n+(-1)^{n-1}}{n}$ 的图象，为了方便看 输出前50项的值

```python
import matplotlib.pyplot as plt
import numpy as np
from fractions import Fraction

# 定义数列
def sequence_fraction(n):
    return Fraction(n + (-1)**(n-1), n)

# 生成 n 值
n_values = np.arange(1, 51)
sequence_fraction_values = [sequence_fraction(n) for n in n_values]

# 输出前50项的值，使用分数表示
for n, val in zip(n_values, sequence_fraction_values):
    print(f"n = {n}: {val}")

# 绘制数列图像
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence_fraction_values, marker='o', linestyle='-', color='b')
plt.title(r'Sequence $\frac{n+(-1)^{n-1}}{n}$')
plt.xlabel('n')
plt.ylabel(r'$\frac{n+(-1)^{n-1}}{n}$')
plt.grid(True)
plt.show();

```
